```python
#Import Pandas Library
import pandas as pd
```

```python
#Import numpy Library
import numpy as np
```

```python
#Import Crime_data dataset
data = pd.read_excel('crim_usa_2016.xls')
```

```python
#View the imported raw
data
```

```python
#Updating the second row as header
New_header = data.iloc[2]
```

```python
#Set the header row as the table header
data.columns = New_header
```

```python
data
```

```python
#Cleaning the unrequired data
data.drop(data.index[:3], inplace=True)
```

```python
data
```

```python
#Removing the NaN calues
data.drop(data.index[20:42], inplace=True)
```

```python
data
```

```python
#Drop columns with NaN values 
data.drop(["Rape\n(revised \ndefinition3)", "Rape\n(revised \ndefinition) \nrate3"], axis=1, inplace = True)
```

```python
data
```

```python
#Resizing the column using pandas library
pd.set_option('max_colwidth', 1000)
```

```python
data
```

```python
##Create a variable that groups and sums up crime rate with population.
crime_rate = data.groupby(['Population1'])['Violent \ncrime \nrate '].sum()
```

```python
##Import of matplot library for making visulazations
import matplotlib.pyplot as mpltb
```

```python
#mpltb.xlabel(" Crime Rate")
#BoxOffice_Income.plot(kind='bar')
crime_rate.plot(kind = "bar", title = "CRIME RATE")
```

```python
#Graph with legend in the top right, x and y labels as well as a title
data.plot(x = "Population1", y = "Violent \ncrime \nrate ", label = "Violent crime rate")
mpltb.xlabel("Population")
mpltb.ylabel("Violent Crime Rate")
mpltb.legend()
mpltb.title("Violent crime rate in relation to population size")
mpltb.show()
```
